### Handling Missing Data

#### Checking the missing values and removing the coloumns which have missing percentage  greater than 50% and  removing the rows which have missing percentage  greater than 50%

```python
#Finding the number of missing values in each column and show the percentage of missing values
print((df.isnull().sum()/len(df)*100).round(2).sort_values(ascending=False))

# A bar plot  of the number of missing values in each column without including the coloumns with 0% missing values
plt.figure(figsize=(8, 6)); 
ax = sns.barplot(x=df.isnull().mean().sort_values(ascending=False)[df.isnull().mean().sort_values(ascending=False) > 0].values*100, y=df.isnull().mean().sort_values(ascending=False)[df.isnull().mean().sort_values(ascending=False) > 0].index); 
plt.xlabel("Percentage of Missing Values");
plt.ylabel("Columns");
plt.title("Percentage of Missing Values in Each Column"); 
[ax.text(value, index, f'{value:.1f}%', color='black', ha="left", va="center") for index, value in enumerate(df.isnull().mean().sort_values(ascending=False)[df.isnull().mean().sort_values(ascending=False) > 0].values*100)];
plt.show()

```

```python
# Removing the columns with more than 50% of missing values
df.dropna(thresh=0.5*len(df), axis=1, inplace=True)

# Checking the number of missing values in each column and show the percentage of missing values
missing=(df.isnull().sum()/len(df)*100).round(2).sort_values(ascending=False)
missing
```

```python
#Finding the missing values in each row and show the percentage of missing values
(df.isnull().sum(axis=1)/len(df)*100).round(2).sort_values(ascending=False)
```

```python
#Finding the data types of the columns with missing percentage greater than 0%
df[missing.index[missing>0]].dtypes
```

```python
# Filter only the float data type columns with missing values (greater than 0%)
float_columns_with_missing = missing[missing > 0].index.intersection(df.select_dtypes(include='float').columns)

# Replace missing values in float columns with the median value
df[float_columns_with_missing] = df[float_columns_with_missing].fillna(df[float_columns_with_missing].median())

# Replace missing values in the "ACCIDENT_TYPE" column with "Unknown"
df["ACCIDENT_TYPE"].fillna("Unknown", inplace=True)
```

### Data Visualization

```python
# Checking the unique values of the columns of the data type 'object'
df.select_dtypes(include='object').apply(pd.Series.nunique, axis=0)
```

```python
# Checking the unique values of the column of Classification
counts = df["CLASSIFICATION"].value_counts()
counts
```

```python
colors = sns.color_palette('pastel')
```

```python
# Calculate total count
total_count = counts.sum()

# Create a subplot
fig, ax = plt.subplots(dpi=300)

# Plot the horizontal bar chart
counts[::-1].plot(kind="barh", color=colors)

plt.xlabel('Classification', fontsize=10)
plt.ylabel('Count', fontsize=10)
plt.title('Counts by Class', fontsize=16)
plt.tick_params(axis='y', labelsize=6)
plt.xlim(right=25000)

# Annotate count values on top of the bars
for i, count in enumerate(counts[::-1]):
    # Calculate percentage value
    percentage = (count / total_count) * 100
    ax.text(count + 0.1, i, f'{percentage:.2f}%', ha='left', va='center',fontsize=6)

plt.show()
```

```python
# Counts of the accidents type 
counts_accident = df["ACCIDENT_TYPE"].value_counts()
counts_accident
```

```python
# Plotting the heatmap of the  accidents type
# Calculate total count
total_count = counts_accident.sum()

# Calculate percentage value for each accident type
percentage_accident = (counts_accident / total_count) * 100

# Create a DataFrame with the percentages only
df_heatmap = pd.DataFrame({'Percentage': percentage_accident})

# Create a heatmap using seaborn
plt.figure(figsize=(10, 10))  # Adjust the figure size as per your preference
sns.heatmap(df_heatmap, annot=True, fmt='.2f', cmap='YlGnBu', cbar=False)

plt.ylabel('Accident Type', fontsize=10)
plt.title('Percentage of Accidents by Type', fontsize=16)

plt.show()
```

```python
#Plot of the accidents type which has a percentage of greater than 1%

# Set a threshold value for the percentages (e.g., 1%)
threshold = 1

# Create a DataFrame with the percentages above the threshold only
df_heatmap = pd.DataFrame({'Percentage': percentage_accident[percentage_accident >= threshold]})

# Find the accident type with the highest percentage above the threshold
max_percentage_accident = df_heatmap.idxmax()[0]


# Create a heatmap using seaborn with the custom color map
plt.figure(figsize=(6, 4))  # Adjust the figure size as per your preference
sns.heatmap(df_heatmap, annot=True, fmt='.2f', cmap='YlGnBu', cbar=False)

plt.ylabel('Accident Type', fontsize=10)
plt.title('Percentage of Accidents by Type', fontsize=16)

plt.show()



```

```python
# Plot for the Degree of injury

counts_degree_injury = df["DEGREE_INJURY"].value_counts()
# Calculate total count for degree of injury
total_count_degree_injury = counts_degree_injury.sum()

# Calculate percentage value for each degree of injury category
percentage_degree_injury = ((counts_degree_injury / total_count_degree_injury) * 100).round(1)

# Create a DataFrame with the percentages for degree of injury
df_degree_injury = pd.DataFrame({'Percentage': percentage_degree_injury})

# Add the degree of injury category as a new column in the DataFrame
df_degree_injury['Degree of Injury'] = df_degree_injury.index

# Create the sunburst chart
fig = go.Figure(go.Sunburst(
    ids=df_degree_injury['Degree of Injury'],
    labels=df_degree_injury['Degree of Injury'] + ' : ' + df_degree_injury['Percentage'].astype(str) + '%',
    parents=[''] * len(df_degree_injury),
    values=df_degree_injury['Percentage'],
))

# Update the layout
fig.update_layout(
    title='Chart of Degree of Injury/Illness with percentage:',
)

fig.show()
```

```python
# Create the treemap plot for the 'Activity' column


# Calculate the value counts for each category in the "ACTIVITY" column
value_counts = df["ACTIVITY"].value_counts()

# Calculate the percentage for each category
percentages = (value_counts / value_counts.sum()) * 100

# Create a DataFrame with the counts and percentages
df_treemap = pd.DataFrame({'Count': value_counts, 'Percentage': percentages})

# Create a temporary variable to denote categories with percentages less than 10% as "Other"
threshold = 3
df_treemap['Activity_Grouped'] = df_treemap.apply(lambda x: 'Other' if x['Percentage'] < threshold else x.name, axis=1)

# Sum the counts for the grouped categories
grouped_counts = df_treemap.groupby('Activity_Grouped')['Count'].sum()

# Create a treemap using squarify
plt.figure(figsize=(10, 6))  # Adjust the figure size as per your preference
colors = sns.color_palette('Paired', n_colors=len(grouped_counts))

squarify.plot(sizes=grouped_counts.values, label=grouped_counts.index, color=colors)

plt.title('Activities during incident', fontsize=16)
plt.axis('off')  # Turn off axis labels


plt.show()
```
