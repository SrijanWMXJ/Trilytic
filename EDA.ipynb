### Handling Missing Data

#### Checking the missing values and removing the coloumns which have missing percentage  greater than 50% and  removing the rows which have missing percentage  greater than 50%

```python
#Finding the number of missing values in each column and show the percentage of missing values
print((df.isnull().sum()/len(df)*100).round(2).sort_values(ascending=False))

# A bar plot  of the number of missing values in each column without including the coloumns with 0% missing values
plt.figure(figsize=(8, 6)); 
ax = sns.barplot(x=df.isnull().mean().sort_values(ascending=False)[df.isnull().mean().sort_values(ascending=False) > 0].values*100, y=df.isnull().mean().sort_values(ascending=False)[df.isnull().mean().sort_values(ascending=False) > 0].index); 
plt.xlabel("Percentage of Missing Values");
plt.ylabel("Columns");
plt.title("Percentage of Missing Values in Each Column"); 
[ax.text(value, index, f'{value:.1f}%', color='black', ha="left", va="center") for index, value in enumerate(df.isnull().mean().sort_values(ascending=False)[df.isnull().mean().sort_values(ascending=False) > 0].values*100)];
plt.show()

```

```python
# Removing the columns with more than 50% of missing values
df.dropna(thresh=0.55*len(df), axis=1, inplace=True)

# Checking the number of missing values in each column and show the percentage of missing values
missing=(df.isnull().sum()/len(df)*100).round(2).sort_values(ascending=False)
missing
```

```python
#Finding the missing values in each row and show the percentage of missing values
(df.isnull().sum(axis=1)/len(df)*100).round(2).sort_values(ascending=False)
```

```python
#Finding the data types of the columns with missing percentage greater than 0%
df[missing.index[missing>0]].dtypes
```

```python
# Filter only the float data type columns with missing values (greater than 0%)
float_columns_with_missing = missing[missing > 0].index.intersection(df.select_dtypes(include='float').columns)

# Replace missing values in float columns with the mode for specified columns
mode_impute_columns = ['UG_MINING_METHOD_CD', 'UG_LOCATION_CD',
                       'OCCUPATION_CD', 'INJ_BODY_PART_CD', 'NATURE_INJURY_CD', 'ACTIVITY_CD', 
                       'INJURY_SOURCE_CD', 'DEGREE_INJURY_CD']
df[mode_impute_columns] = df[mode_impute_columns].fillna(df[mode_impute_columns].mode().iloc[0])

# Replace missing values in float columns with the median for specified columns
median_impute_columns = ['SCHEDULE_CHARGE', 'DAYS_RESTRICT', 'DAYS_LOST', 'TOT_EXPER', 'MINE_EXPER', 
                         'JOB_EXPER', 'SHIFT_BEGIN_TIME']
df[median_impute_columns] = df[median_impute_columns].fillna(df[median_impute_columns].median())
```

```python
import spacy
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from scipy.sparse import hstack

nlp = spacy.load("en_core_web_sm")

# Dataframe df with the column narrative containing the sentence
df["tokens"] = df["NARRATIVE"].apply(lambda text: [token.text for token in nlp(text) if not token.is_stop])
```

```python
# Convert the tokenized text back to sentences (optional)
df["processed_narrative"] = df["tokens"].apply(lambda tokens: ' '.join(tokens))
vectorizer = TfidfVectorizer()
# Separate features (tokenized text) and target (accident classification)
X_text = df["processed_narrative"]
X_text_vec = vectorizer.fit_transform(X_text)

#UG_LOCATION, UG_MINING_METHOD, UG_MINING_METHOD, MINING_EQUIP, EQUIP_MFR_NAME,IMMED_NOTIFY_CD,IMMED_NOTIFY

# Select multiple categorical columns
categorical_columns = df.drop(["NARRATIVE","IMMED_NOTIFY","IMMED_NOTIFY_CD","EQUIP_MFR_NAME","MINING_EQUIP","UG_MINING_METHOD","UG_MINING_METHOD","UG_LOCATION","processed_narrative","tokens","CLASSIFICATION"],axis =1).select_dtypes(include='object').columns  # Replace with your actual column names
print(categorical_columns)
# One-hot encode the categorical columns
encoder = OneHotEncoder(handle_unknown="ignore")
X_categorical = encoder.fit_transform(df[categorical_columns])

# Combine text and categorical features
from scipy.sparse import hstack
X_combined = hstack((X_text_vec, X_categorical))

# Separate target variable
y = df["CLASSIFICATION"]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, random_state=42)

# Train the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print(classification_report(y_test, y_pred))
```

```python
# Print confusion matrix
print("\nConfusion Matrix:")
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)

row_sums = conf_matrix.sum(axis=1, keepdims=True)
norm_conf_matrix = conf_matrix / row_sums

# Display colored confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(norm_conf_matrix, annot=True, xticklabels=rf_model.classes_, yticklabels=rf_model.classes_)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```
